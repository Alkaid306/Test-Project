```python
# 统计每一列NaN的数量

df.isnull().sum().sort_values(ascending=False)
```

```python
# 统计train,test所有列中的：数值类型的列
number_columns = [ col for col in df.columns if df[col].dtype != 'object']
```

```python
# 统计 train中有哪些列包含NaN
train_nan_num = [] # train中数值类型的列
for col in number_columns:
    if df[col].isnull().sum() > 0:
        train_nan_num.append(col)
```

```python
len(train_nan_num)
```

```python
# 将空缺值用中位数替代
for col in train_nan_num:
    df[col].fillna(df[col].median(), inplace=True) # 中位数替代
```

```python
# 过滤低方差法
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.5)     
X_vt = pd.DataFrame(vt.fit_transform(df),columns=df.columns[vt.get_support()] )     
X_vt.head()
```

```python
# 归一化所有数据
from sklearn.preprocessing import StandardScaler

df['一天去两家医院的天数_'] = StandardScaler().fit_transform(df['一天去两家医院的天数'].values.reshape(-1,1)) # 金额归一化
df.drop(['一天去两家医院的天数'], axis=1, inplace=True) # 删除原始的数据列

df['就诊的月数_'] = StandardScaler().fit_transform(df['就诊的月数'].values.reshape(-1,1)) # 金额归一化
df.drop(['就诊的月数'], axis=1, inplace=True) # 删除原始的数据列

df['月就诊天数_MAX_'] = StandardScaler().fit_transform(df['月就诊天数_MAX'].values.reshape(-1,1)) # 金额归一化
df.drop(['月就诊天数_MAX'], axis=1, inplace=True) # 删除原始的数据列

df['月就诊天数_AVG_'] = StandardScaler().fit_transform(df['月就诊天数_AVG'].values.reshape(-1,1)) # 金额归一化
df.drop(['月就诊天数_AVG'], axis=1, inplace=True) # 删除原始的数据列

df['月就诊医院数_MAX_'] = StandardScaler().fit_transform(df['月就诊医院数_MAX'].values.reshape(-1,1)) # 金额归一化
df.drop(['月就诊医院数_MAX'], axis=1, inplace=True) # 删除原始的数据列
```
